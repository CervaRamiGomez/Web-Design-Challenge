```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
cities_data_path = "Resources/cities-2.csv"


# Read the mouse data and the study results
cities_data_path = pd.read_csv(cities_data_path)

# Display the data table for preview
cities_data_path
```

```python

# name X & Y values
x = cities_data_path['Lat']
y = cities_data_path['Max Temp']

# Create a scatter plot based on new dataframe above with circle markers and listed colors
plt.scatter(x, y)

# Add labels and title to plot
plt.xlabel("Latitude")
plt.ylabel("Max Tempurature (F)")
plt.title('City Latitude Vs Max Tempurature 01/05/17')
# Display plot
plt.show()
```
